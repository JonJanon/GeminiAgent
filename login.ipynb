```python
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from email.message import EmailMessage
import base64
import os

# Define the SCOPES and CLIENT_SECRET_FILE path
SCOPES = ['https://www.googleapis.com/auth/gmail.readonly', 'https://www.googleapis.com/auth/gmail.compose']
CLIENT_SECRET_FILE = 'credentials2.json'
TOKEN_FILE = 'token.json'
from_name=""
signature="test_signature"
to_address=""
subject=""
def authenticate_gmail():
    # Check if token file exists
    if os.path.exists(TOKEN_FILE):
        credentials = Credentials.from_authorized_user_file(TOKEN_FILE, SCOPES)
    else:
        flow = InstalledAppFlow.from_client_secrets_file(CLIENT_SECRET_FILE, SCOPES)
        credentials = flow.run_local_server(port=0)

        # Save the credentials for the next run
        with open(TOKEN_FILE, 'w') as token:
            token.write(credentials.to_json())

    return credentials

def get_past_emails():
    return 0

# to test the api, try to get the most recent email
def get_most_recent_email():
    
    # Load credentials from the saved token file
    credentials = authenticate_gmail()

    # Build the Gmail API service
    service = build('gmail', 'v1', credentials=credentials)

    # Call the Gmail API to get the list of unread messages
    #results = service.users().messages().list(userId='me', q='is:unread').execute()
    results = service.users().messages().list(userId='me',labelIds=['INBOX'], maxResults=1).execute()
    messages = results.get('messages', [])

    if not messages:
        print('No unread messages found.')
        return

    # Get the ID of the most recent unread message
    most_recent_message_id = messages[0]['id']

    # Retrieve the message content
    message = service.users().messages().get(userId='me', id=most_recent_message_id).execute()
    return message

if __name__ == '__main__':
    # Get the most recent unread email and respond
    print(get_most_recent_email())
```
